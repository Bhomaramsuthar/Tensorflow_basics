```python
def HelloWorldXY(x,y):
    if(x<10):
        print("Hello World , x was < 10")
    elif(x<20):
        print("Hello World, x was >= 10 but < 20")
    else:
        print("Hello World , x was >=20")
    return x+y

for i in range(8, 25, 5):
    print("--- Now running with i: {}".format(i))
    r = HelloWorldXY(i,i)
    print("Result from HelloWorld: {}".format(r))
```

```python
print(HelloWorldXY(11,2))
```

```python
print("Iterate over the item. `range(2)` is like a list [0,1].")
for i in range(2):
    print(i)

print("Iterate over an actual list.")
for i in [0,1]:
    print(i)

print("While works")
i=0
while i < 2:
    print(i)
    i += 1
```

# Numpy and List

Here enumerate add the counter to the iterator .<br> '.format()' is used to replace the llaceholder placed into string with the provided value

```python
import numpy as np

a = np.array(["Hello","World"])
a = np.append(a,"!")
print("Current array:{}".format(a))
print("Printing each element")
for i in a:
    print(i)

print("\n Printing each element with index")
for i,e in enumerate(a):
    print("Index : {} and Element : {}".format(i,e))

```

```python
print("Basic maths in numpy")
x = np.array([0,1,2,3,4,5])
print("The max is {}".format(np.max(x)))
print("The average is {}".format(np.average(x)))
print("The index of max value {}".format(np.argmax(x)))
```

```python
print("Creating a random matrix of 3*3 with random values")
y = np.random.rand(3,3)
print(y)
```

```python
print("Dimensions of array")
print("Shapes of a:{}".format(a.shape))
print("Shapes of x:{}".format(x.shape))
print("Shapes of y:{}".format(y.shape))
```
