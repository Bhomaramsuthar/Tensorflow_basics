```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
import numpy as np
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)
```

# Training data

```python
c_temperature = np.array([0,10,7,85,12,-15], dtype=float)
f_temperature = np.array([32,50,44.6,185,53.6,5], dtype=float)

for i,c in enumerate(c_temperature):
    print("{} degrees celcius = {} degrees fahrenheit".format(c,f_temperature[i]))
```

# Plotting the data

```python
plt.scatter(c_temperature,f_temperature, label="Data", color="blue")
plt.title("c VS f temperature")
plt.xlabel("Temperature in C")
plt.ylabel("Temperature in F")
plt.legend()
plt.grid(True)
plt.show()
```

# Predefining layers the adding to the model

here l0 = layer0

```python
l0 = tf.keras.layers.Dense(units=1,input_shape=[1])
```

```python
model = tf.keras.Sequential([l0])
```

# Defining layers inside the model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1,input_shape=[1])
])
```

# Compiling model
commpile = loss + optimizer

```python
model.compile(
    loss="mean_squared_error",
    optimizer=tf.keras.optimizers.Adam(0.1)
)
```

# Training

```python
history = model.fit(c_temperature,f_temperature,epochs=500,verbose=False)
print("Finished training the model")
```

```python
plt.xlabel("Epoch Number")
plt.ylabel("Loss magnitude")
plt.plot(history.history['loss'])
```

# Prediction

predict 100 degree C to 212 degree F

```python
print(model.predict([100]))
```

close it's good

```python
print("Layers weights {}".format(l0.get_weights()))
print("Model weights {}".format(model.get_weights()))
```

```python
pred_val = np.array([1,2,3,4], dtype=float).reshape(-1,1)
valid_val = np.array([33.8,35.6,37.4,39.2], dtype=float)
model_pred=model.predict(pred_val)
print("Model predicted {}".format(model_pred))
```

```python
plt.scatter(pred_val,valid_val,label='True values' ,color='blue')
plt.scatter(pred_val,model_pred,label='Predict values' ,color='red')
plt.title("c VS f temperature")
plt.xlabel("C in temperature")
plt.ylabel("F in temperature")
plt.legend()
plt.grid(True)
plt.show()
```
